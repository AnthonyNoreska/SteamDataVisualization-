```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import matplotlib.pyplot as plt
import pandas as pd
import time
import requests
import json
```

```python
csv_files = [
    "Action_data_converted.csv",
    "Adventure_data_converted.csv",
    "Indie_data_converted.csv",
    "RPG_data_converted.csv",
    "Simulation_data.csv",
    "Simulation_data_converted.csv",
    "Sports_data_converted.csv",
    "Strategy_data_converted.csv"
]

# Create a dictionary to store the dataframes
dataframes = {}

# Loop through the CSV files and read them into dataframes
for file in csv_files:
    # Extract the name of the dataframe from the file name
    df_name = file.split('.')[0]
    # Read the CSV file into a dataframe
    dataframes[df_name] = pd.read_csv(file)
```

```python
df=dataframes['Action_data_converted']
```

```python
import pandas as pd

# Create a new column 'rating_percentage' based on the percentage of positive reviews
df['rating_percentage'] = (df['positive'] / (df['positive'] + df['negative'])) * 100

# Display the dataframe with the new 'rating_percentage' column
print(df[['name', 'rating_percentage']])
```

```python
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'price' and 'ccu' are columns in your dataset
prices = df['price']
ccu = df['ccu']

# Convert 'price' column to string and then to numeric (remove non-numeric characters)
prices = pd.to_numeric(prices.astype(str).str.replace('[^\d.]', ''), errors='coerce')

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(prices, ccu, alpha=0.5)
plt.title('Price vs Concurrent Users')
plt.xlabel('Price')
plt.ylabel('Concurrent Users')
plt.grid(True)
plt.show()
```
